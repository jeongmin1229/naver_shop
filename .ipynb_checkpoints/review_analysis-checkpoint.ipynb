```python
from konlpy.tag import Kkma, Okt, Komoran
kkma = Kkma()
twitter = Okt()
komoran = Komoran()
```

```python
line_list = []


with open('test_all.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()
    lines = [line.strip() for line in lines]
    lines[:] = [x for x in lines if x]
#     lines.remove('.')
#     print(lines)
    for line in lines:
#         print(line)
        line = kkma.nouns(line)
        line_list.append(line)
```
